```python
import makeitseg2, os
```

```python
def seperateSUfiles(folderLocation):
    files = os.listdir(folderLocation)
    files = [eachFile for eachFile in files if eachFile.endswith('rvz.su')]
    return files
```

```python
folderLocation = 'seismic_files'
folderTypes = os.listdir(folderLocation)
```

```python
for eachFolderType in folderTypes:
    folders = os.listdir(folderLocation + '/' + eachFolderType)
    for eachFolder in folders:
        filesLocation = folderLocation + '/' + eachFolderType + '/' + eachFolder
        files = seperateSUfiles(filesLocation)
        if len(files):
            count=1
            for file in files:
                print(count,file)
                fileLocation = filesLocation + '/' + file
                !surange < $fileLocation
                new_name_num = []
                for char in file:
                    if char.isdigit():
                        new_name_num.append(char)
                new_name = ''.join(new_name_num)
                image_name_eps = f'{filesLocation}/image{new_name}.eps'
                !supswigb < $fileLocation perc=99 > $image_name_eps
                image_name_jpg = f'{filesLocation}/image{new_name}.jpg'
                print(image_name_eps, ' saved')
                !convert $image_name_eps $image_name_jpg
                s1 = new_name                                               
                s2 = "0"
                if s1.startswith(s2) and s1 != "0":
                    s3 = s1.replace(s2, '')
                    new_name = s3
                if not os.path.exists(filesLocation+"/cut_su/"):
                    os.mkdir(filesLocation+"/cut_su/")
                
                print("Number of the shot is  : ", new_name)
                sx = int(new_name)
                new_name_full = filesLocation+"/cut_su/"+"fault" + new_name + ".su"
                print("New name is : ", new_name_full)
                                                  
                min_tracl_loc = sx
                gx = 20
                min_tracl = int(min_tracl_loc/gx)+2
                print("min_tracl : ", min_tracl)                   
                if min_tracl < 0 :
                    min_tracl = 0
                !suwind < $fileLocation key=tracl min=$min_tracl > $new_name_full 
                !surange < $new_name_full
                count+=1
                
            for file_name in os.listdir(filesLocation):
                seg2fname = f"{filesLocation}/cut_su/{file_name.split('/')[-1].split('.')[0]}.dat"
                if file_name.endswith('rvz.su'):
                    makeitseg2.convert(new_name_full, seg2fname)
```
